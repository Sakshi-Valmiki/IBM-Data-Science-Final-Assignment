```python
# Import yfinance
import yfinance as yf

# Create a ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract stock data
tesla_data = tesla.history(period="max")

# Reset the index
tesla_data.reset_index(inplace=True)

# Display the first five rows
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Download the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text

# Parse the HTML
soup = BeautifulSoup(html_data, "html.parser")

# Create an empty DataFrame
tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Locate the Tesla revenue table
table = soup.find_all("tbody")[1]

# Extract the data
for row in table.find_all("tr"):
    cols = row.find_all("td")

    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()

        tesla_revenue = pd.concat(
            [
                tesla_revenue,
                pd.DataFrame({"Date": [date], "Revenue": [revenue]})
            ],
            ignore_index=True,
        )

# Remove commas and dollar signs
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(",", "", regex=False)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "", regex=False)

# Remove null and empty values
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]

# Display the last five rows
tesla_revenue.tail()
```

```python
# Import yfinance
import yfinance as yf

# Create a ticker object for GameStop
gme = yf.Ticker("GME")

# Extract GameStop stock data
gme_data = gme.history(period="max")

# Reset the index
gme_data.reset_index(inplace=True)

# Display the first five rows
gme_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Download the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data_2 = requests.get(url).text

# Parse the HTML
soup = BeautifulSoup(html_data_2, "html.parser")

# Create an empty DataFrame
gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])

# Find the GameStop Revenue table
table = soup.find_all("tbody")[1]

# Extract the data
for row in table.find_all("tr"):
    cols = row.find_all("td")

    if len(cols) == 2:
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()

        gme_revenue = pd.concat(
            [
                gme_revenue,
                pd.DataFrame({"Date": [date], "Revenue": [revenue]})
            ],
            ignore_index=True
        )

# Clean the Revenue column
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace("$", "", regex=False)
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(",", "", regex=False)

# Remove null and empty values
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]

# Display the last five rows
gme_revenue.tail()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert Date columns to datetime
tesla_data["Date"] = pd.to_datetime(tesla_data["Date"])
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"])

# Convert Revenue to numeric
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(str)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(",", "", regex=False)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace("$", "", regex=False)
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"])

# Keep data only up to June 2021
tesla_stock = tesla_data[tesla_data["Date"] <= "2021-06-30"]
tesla_rev = tesla_revenue[tesla_revenue["Date"] <= "2021-06-30"]

# Create graph
fig, ax1 = plt.subplots(figsize=(14,7))

ax1.plot(tesla_stock["Date"], tesla_stock["Close"], color="blue")
ax1.set_xlabel("Date")
ax1.set_ylabel("Stock Price (USD)", color="blue")

ax2 = ax1.twinx()
ax2.plot(tesla_rev["Date"], tesla_rev["Revenue"], color="red")
ax2.set_ylabel("Revenue (Millions USD)", color="red")

plt.title("Tesla Stock Price and Revenue")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert Date columns to datetime
gme_data["Date"] = pd.to_datetime(gme_data["Date"])
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"])

# Clean and convert Revenue column to numeric
gme_revenue["Revenue"] = gme_revenue["Revenue"].astype(str)
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(",", "", regex=False)
gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace("$", "", regex=False)
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]
gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"])

# Keep data only up to June 2021
gme_stock = gme_data[gme_data["Date"] <= "2021-06-30"]
gme_rev = gme_revenue[gme_revenue["Date"] <= "2021-06-30"]

# Create graph
fig, ax1 = plt.subplots(figsize=(14,7))

# Plot stock price
ax1.plot(gme_stock["Date"], gme_stock["Close"], color="blue")
ax1.set_xlabel("Date")
ax1.set_ylabel("Stock Price (USD)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Plot revenue
ax2 = ax1.twinx()
ax2.plot(gme_rev["Date"], gme_rev["Revenue"], color="red")
ax2.set_ylabel("Revenue (Millions USD)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Graph title
plt.title("GameStop Stock Price and Revenue")

plt.show()
```
